## Q1 TF-IDF (50 Points)

Recall that the TF-IDF score of a term ($w$) in a document ($d$) is calculated as:

> $\text{TF-IDF}_{w,d} = TF_{w,d} \cdot \log (IDF_w)$

Where:

- $N_d$ is the total number of documents.
- $TF_{w,d}$ is the **Term Frequency** of word $w$ in document $d$, defined as the count of $w$ in $d$.
- $DF_w$ is the **Document Frequency** of word $w$, i.e., the number of documents that contain $w$, divided by the total number of documents.
- $IDF_w$ is the **Inverse Document Frequency** of word $w$. That is, $\frac{1}{DF}$

### Q1.2 Normalized TF

**Please read the following before answering Q1.2:**

In practice, we typically normalize term frequency ($TF_{w,d}$) by dividing it by the total number of tokens in the document $d$. This ensures that the sum of term frequencies in each document (i.e., each row) always equals one. By doing so, we prevent longer documents from having higher raw term frequencies.

For example, consider the following Document-Term Matrix:

|   | A   | B   |
|---|-----|-----|
| D1 | 1   | 1   |
| D2 | 100 | 100 |

In **D1**, we have one occurrence of **A** and one of **B**. In **D2**, we have 100 occurrences of **A** and 100 of **B**. Clearly, **D2** is much longer than **D1**. If we naively use the raw term frequencies from the table, a machine learning model might assign more importance to **D2** simply because its term frequencies are larger. However, **D1** and **D2** are fundamentally the same, as both terms **A** and **B** are **equally distributed** in them.

To address this issue, we normalize the term frequencies by dividing each count by the total number of tokens in the document. This transforms the table into:

|   | A         | B         |
|---|----------|----------|
| D1 | 1/2 = 0.5 | 1/2 = 0.5  |
| D2 | 100/200 = 0.5 | 100/200 = 0.5 |

Now, both **D1** and **D2** have the same representation, ensuring that document length does not distort term importance.
