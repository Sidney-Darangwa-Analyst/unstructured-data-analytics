```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import gdown
from google.colab.patches import cv2_imshow
## Imports may take a few minutes

import ultralytics
from deepface import DeepFace
from ultralytics import YOLO
```

```python
# helper function to plot the image from a path
def plot_image(img_path, max_height=400):
    '''A helper function to plot the image from a path.

    Args:
        img_path (str): The path to the image.
    '''
    # get the image from url
    img = cv2.imread(img_path)

    # Get original dimensions
    h, w = img.shape[:2]

    # Resize only if height exceeds max
    if h > max_height:
        scale = max_height / h
        new_w = int(w * scale)
        new_h = int(h * scale)
        resized_img = cv2.resize(img, (new_w, new_h))
    else:
        resized_img = img  # No resizing needed

    # Show image
    cv2_imshow(resized_img)
```

## Facial Analysis

### Face Detection

Let's detect the face from an image.

```python
# let's plot the original image
plot_image('keanu.jpg')
```

```python
# extract the face from the image
result = DeepFace.extract_faces(
    img_path='keanu.jpg',
    detector_backend='mtcnn',  # choose your model! mtcnn, retinaface, dlib, yolo, yunet
    align=False,
)
```

The result is a list, each element contains the face image and the facial area coordinates.

```python
# print the length of the result
len(result)
```

```python
# let's print out the facial area coordinates
result[0]['facial_area']
```

```python
# let's print out the detection confidence
result[0]['confidence']
```

```python
# let's print out the content of the result
# you'll find it has three keys: 'face', 'facial_area', and 'confidence'
result[0]['face']
```

```python
# print the detected (cropped) face
plt.imshow(result[0]['face'])
plt.axis('off')
plt.show()
```
