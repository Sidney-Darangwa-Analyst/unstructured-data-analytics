#### Q1.3.1 Calculating User Similarity

To find users with similar tastes to Eva, we need to calculate similarity scores between users. Let's walk through this process:

The first step for user-based filtering is to calculate the similarity between Eva and other users. For example, to calculate the similarity between Eva and Andy, we first extract the movies both Eva and Bob have rated: "Arrival, Hail Ceasar", "Rogue One", and "The Jungle Book". Next, we construct two vectors that represent Eva and Bob. For Eva, the vector is $[2, 4, 3, 5]$.

<font color = orange > Please construct the vector for Andy and calculate the cosine similarity between **Eva** and **Andy**. </font>

<font color = orange > You may use `numpy`'s cosine similarity function. </font>

---
FYI: The cosine similarity is defined as

$$
\text{cosine_similarity}(\mathbf{A}, \mathbf{B}) = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}
$$


Where:
- $\mathbf{A} \cdot \mathbf{B}$ is the dot product of the vectors
- $\|\mathbf{A}\|$ and $\|\mathbf{B}\|$ are the magnitudes (Euclidean norms) of the vectors

As an example, let

$$
\mathbf{A} = [3, 4], \quad \mathbf{B} = [4, 3]
$$

1. **Dot product**:

$$
\mathbf{A} \cdot \mathbf{B} = 3 \cdot 4 + 4 \cdot 3 = 12 + 12 = 24
$$

2. **Magnitudes**:

$$
\|\mathbf{A}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

$$
\|\mathbf{B}\| = \sqrt{4^2 + 3^2} = \sqrt{16 + 9} = \sqrt{25} = 5
$$

3. **Cosine similarity**:

$$
\text{cosine_similarity}(\mathbf{A}, \mathbf{B}) = \frac{24}{5 \cdot 5} = \frac{24}{25} = 0.96
$$
